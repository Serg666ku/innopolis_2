```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df3=pd.read_csv ('df3.csv')
print(len(df3))
print(df3.head())
```

Эти данные представляют собой заводские производственные показатели и сообщенное количество дефектов в определенные дни недели.

**1. Воссоздайте этот график разброса «произведенных» и «дефектных»  деталей. Обратите внимание на цвет и размер точек. Также обратите внимание на размер картинки. Сможете ли растянуть его аналогичным образом?**

```python
df3.plot.scatter('produced','defective',color = 'g', figsize=(13,5))
```

**2. Создайте гистограмму для колонки 'produced'**

```python
df3['produced'].plot.hist()
```

**3. Перерисуйте гистограмму как на рисунке ниже**

```python
df3[['produced']].plot.hist(bins = 10,color='g',edgecolor ='w',legend = None).autoscale(enable =True, axis = 'x', tight = True)
```

**4. Создайте boxplot для 'produced' в каждый из дней недели 'weekday' (используйте группировку)**

```python
df3[['produced','weekday']].boxplot(by ='weekday', grid = True)
```

**5.Создайте KDE (ядерная оценка плотности) для колонки 'defective'**

```python
df3['defective'].plot.kde(color='r')
```

**6. Видоизмените график KDE, как на рисунке ниже**

```python
df3['defective'].plot.kde(linestyle ='--', linewidth = '4',color = 'g')
```

**7. Для первых 30 строк постройте графики областей «произведенных» и «дефектных» деалей. (используйте .loc)**

```python
df3.loc[:30][['produced','defective']].plot.area(colormap ='coolwarm', alpha = 0.5, stacked = False)
```

```python
#поменяйте расположение легенды
df3.loc[:30][['produced','defective']].plot.area(colormap ='coolwarm', alpha = 0.5, stacked = False).legend(bbox_to_anchor=(1.4, 0.5))
```
